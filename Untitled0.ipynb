```python
print("=" * 60)
print("ACT 5: POLICY SIMULATION")
print("=" * 60)

R = 2  # Your value

# Current state
current = {
    'trips': len(df_clean),
    'revenue': df_clean['fare_amount'].sum(),
    'avg_fare': df_clean['fare_amount'].mean(),
    'avg_distance': df_clean['trip_distance'].mean()
}

print(f"\n📊 CURRENT STATE (Baseline):")
print(f"Total Trips:     {current['trips']:,}")
print(f"Total Revenue:   ${current['revenue']:,.2f}")
print(f"Average Fare:    ${current['avg_fare']:.2f}")
print(f"Average Distance: {current['avg_distance']:.2f} miles")

# Simulate fare increase by R% (2%)
new_avg_fare = current['avg_fare'] * (1 + R/100)

print(f"\n🔧 SIMULATION: Base Fare +{R}%")
print(f"New Average Fare: ${new_avg_fare:.2f}")

# Different demand elasticity scenarios
# Elasticity = % change in demand per % change in price
scenarios = {
    'Low Elasticity (Inelastic)': -0.3,      # People keep taking taxis
    'Medium Elasticity': -1.5,               # Some people switch
    'High Elasticity (Elastic)': -3.0        # Many people switch
}

print(f"\n{'='*60}")
print("SCENARIO RESULTS")
print(f"{'='*60}")

results = []
for name, elasticity in scenarios.items():
    # Calculate demand change
    demand_change_pct = elasticity * R  # e.g., -1.5 * 2 = -3%

    # New trip count
    new_trips = int(current['trips'] * (1 + demand_change_pct/100))

    # New revenue
    new_revenue = new_trips * new_avg_fare

    # Changes
    trip_change_pct = (new_trips - current['trips']) / current['trips'] * 100
    revenue_change_pct = (new_revenue - current['revenue']) / current['revenue'] * 100

    results.append({
        'Scenario': name,
        'Elasticity': elasticity,
        'Demand Change %': demand_change_pct,
        'New Trips': new_trips,
        'Trip Change %': trip_change_pct,
        'New Revenue': new_revenue,
        'Revenue Change %': revenue_change_pct
    })

    print(f"\n{name}:")
    print(f"  Elasticity: {elasticity}")
    print(f"  Demand change: {demand_change_pct:.1f}%")
    print(f"  New trips: {new_trips:,} ({trip_change_pct:+.1f}%)")
    print(f"  New revenue: ${new_revenue:,.2f} ({revenue_change_pct:+.1f}%)")

# Create results DataFrame
results_df = pd.DataFrame(results)
print(f"\n{'='*60}")
print("SUMMARY TABLE")
print(f"{'='*60}")
print(results_df[['Scenario', 'Demand Change %', 'Trip Change %', 'Revenue Change %']].to_string(index=False))
```

```python
print("=" * 60)
print("ACT 3: HYPOTHESIS TESTING")
print("=" * 60)

# HYPOTHESIS: Weekend trips have longer distances than weekday trips

weekend_dist = df_clean[df_clean['is_weekend']]['trip_distance']
weekday_dist = df_clean[~df_clean['is_weekend']]['trip_distance']

print(f"\n📊 WEEKEND vs WEEKDAY TRIP DISTANCE")
print(f"Weekend count: {len(weekend_dist):,}")
print(f"Weekday count: {len(weekday_dist):,}")
print(f"\nWeekend average: {weekend_dist.mean():.3f} miles")
print(f"Weekday average: {weekday_dist.mean():.3f} miles")

# T-test
t_stat, p_value = stats.ttest_ind(weekend_dist, weekday_dist)

print(f"\n📈 T-TEST RESULTS:")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.2e}")

alpha = 0.05
if p_value < alpha:
    print(f"\n✅ SIGNIFICANT DIFFERENCE (p < {alpha})")
    print("→ Weekend trips ARE significantly longer!")
else:
    print(f"\n❌ NO SIGNIFICANT DIFFERENCE (p >= {alpha})")
    print("→ Weekend trips are NOT significantly longer")

# Effect size (Cohen's d)
pooled_std = np.sqrt(((len(weekend_dist)-1)*weekend_dist.var() +
                       (len(weekday_dist)-1)*weekday_dist.var()) /
                      (len(weekend_dist) + len(weekday_dist) - 2))
cohens_d = (weekend_dist.mean() - weekday_dist.mean()) / pooled_std
print(f"\nEffect size (Cohen's d): {cohens_d:.4f}")
if abs(cohens_d) < 0.2:
    print("→ Negligible effect")
elif abs(cohens_d) < 0.5:
    print("→ Small effect")
elif abs(cohens_d) < 0.8:
    print("→ Medium effect")
else:
    print("→ Large effect")
```

```python
print("=" * 60)
print("ACT 2: CLEANING DATA - DEFINING 'REAL TRIPS'")
print("=" * 60)

# Make a copy to work with
clean = df.copy()

# Convert datetime columns properly
clean['tpep_pickup_datetime'] = pd.to_datetime(clean['tpep_pickup_datetime'])
clean['tpep_dropoff_datetime'] = pd.to_datetime(clean['tpep_dropoff_datetime'])

original_count = len(clean)
print(f"\n📊 Original rows: {original_count:,}")

# ============================================
# RULE 1: Fix crazy dates
# ============================================
# Keep only 2024 dates (remove 2002, future dates, etc.)
clean = clean[clean['tpep_pickup_datetime'].dt.year == 2024]
clean = clean[clean['tpep_dropoff_datetime'].dt.year == 2024]
print(f"After fixing dates: {len(clean):,} ({len(clean)/original_count*100:.1f}%)")

# ============================================
# RULE 2: Remove missing critical data
# ============================================
# We NEED these to analyze: pickup time, distance, fare, passengers
critical_cols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime',
                  'trip_distance', 'fare_amount', 'passenger_count']
clean = clean.dropna(subset=critical_cols)
print(f"After removing missing critical data: {len(clean):,} ({len(clean)/original_count*100:.1f}%)")

# ============================================
# RULE 3: Remove negative and zero fares
# ============================================
# Why? Negative fare = impossible. Zero fare = likely error
clean = clean[clean['fare_amount'] > 0]
print(f"After removing bad fares: {len(clean):,} ({len(clean)/original_count*100:.1f}%)")

# ============================================
# RULE 4: Remove trips with 0 passengers
# ============================================
# Why? Every real trip needs at least 1 person
clean = clean[clean['passenger_count'] > 0]
print(f"After removing 0-passenger trips: {len(clean):,} ({len(clean)/original_count*100:.1f}%)")

# ============================================
# RULE 5: Calculate trip duration
# ============================================
clean['duration_min'] = (
    clean['tpep_dropoff_datetime'] - clean['tpep_pickup_datetime']
).dt.total_seconds() / 60

# ============================================
# RULE 6: Remove unrealistic durations
# ============================================
# Too short (< 1 min) = cancelled or error
# Too long (> 3 hours) = driver forgot to close meter
clean = clean[(clean['duration_min'] >= 1) & (clean['duration_min'] <= 180)]
print(f"After removing bad durations: {len(clean):,} ({len(clean)/original_count*100:.1f}%)")

# ============================================
# RULE 7: Remove 0-distance trips with high fare
# ============================================
# Why? If no distance but high fare = broken meter
clean = clean[~((clean['trip_distance'] == 0) & (clean['fare_amount'] > 10))]
print(f"After removing 0-distance expensive trips: {len(clean):,} ({len(clean)/original_count*100:.1f}%)")

# ============================================
# RULE 8: Keep only January and March
# ============================================
# YOUR assigned months!
clean['month'] = clean['tpep_pickup_datetime'].dt.month
clean = clean[clean['month'].isin([1, 3])]  # January=1, March=3
print(f"After keeping only Jan & Mar: {len(clean):,} ({len(clean)/original_count*100:.1f}%)")

# ============================================
# RULE 9: Remove extreme outliers
# ============================================
# Fares > $500 = likely data entry error
clean = clean[clean['fare_amount'] <= 500]
# Trips > 100 miles in NYC = unrealistic
clean = clean[clean['trip_distance'] <= 100]
print(f"After removing extreme outliers: {len(clean):,} ({len(clean)/original_count*100:.1f}%)")

# ============================================
# FINAL SUMMARY
# ============================================
print("\n" + "=" * 60)
print("✅ CLEANING COMPLETE!")
print("=" * 60)
print(f"Original rows:      {original_count:,}")
print(f"Final clean rows:   {len(clean):,}")
print(f"Rows removed:       {original_count - len(clean):,}")
print(f"Retention rate:     {len(clean)/original_count*100:.1f}%")
print(f"Data lost:          {(original_count - len(clean))/original_count*100:.1f}%")
print("=" * 60)

# Save as your clean dataset
df_clean = clean.copy()
print("\n✅ df_clean is ready for analysis!")
```
